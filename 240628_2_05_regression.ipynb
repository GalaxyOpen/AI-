1. X와 Y에 대한 샘플 데이터를 생성
2. LinearRegression 객체를 인스턴스화
3. 'fit' 방법을 사용하여 데이터에 모델을 적합시킴
4. 다음 회귀선의 기울기(회귀율)와 절편(절편)을 출력


```python
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Sample data
X = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))  # Independent variable
Y = np.array([5, 20, 14, 32, 22, 38])  # Dependent variable
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Fit the model using the training data
model.fit(X, Y)
```

```python
# Print the regression coefficient and intercept
print('회귀 계수 (Slope):', model.coef_[0])
print('절편 (Intercept):', model.intercept_)

# Predict using the model
Y_pred = model.predict(X)
```

```python
# Plotting the results
plt.scatter(X, Y, color='blue')  # actual values
plt.plot(X, Y_pred, color='red')  # regression line
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
