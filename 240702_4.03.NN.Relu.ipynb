```python
import numpy as np
import matplotlib.pyplot as plt

# ReLU 활성화 함수 정의
def relu(x):
    return np.maximum(0, x)

# 예시 데이터 생성
x = np.linspace(-10, 10, 100)
y = relu(x)

# 활성화 함수의 결과를 시각화
plt.plot(x, y)
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```
